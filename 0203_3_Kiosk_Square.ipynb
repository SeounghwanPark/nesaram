```python
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

class Kiosk:
    def __init__(self):
        self.menu = menu  # 메뉴 리스트
        self.price = price  # 가격 리스트
        self.price_sum = 0  # 선택한 음료의 가격(초기값)
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트

    def menu_print(self):
        """메뉴 출력"""
        print("\n 메뉴판")
        for i in range(len(self.menu)):
            print(f"{i+1}. {self.menu[i]} : {self.price[i]}원")

    def menu_select(self):
        """음료 선택 및 주문 추가"""
        while True:
            print("\n 음료를 선택하세요 (0 입력 시 주문 종료)")
            try:
                n = int(input("음료의 번호를 입력하세요 : "))
            except ValueError:
                print(" 숫자로 입력해주세요.\n")
                continue

            # 주문 종료
            if n == 0:
                print("\n 주문이 완료되었습니다.")
                break

            # 유효한 메뉴 선택 확인
            if 1 <= n <= len(self.menu):
                selected_menu = self.menu[n-1]
                selected_price = self.price[n-1]
            else:
                print(" 없는 메뉴입니다. 다시 주문해 주세요.")
                continue

            # 음료 온도 선택
            t = 0
            while t not in [1, 2]:
                try:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = "HOT"
                    elif t == 2:
                        temp = "ICE"
                    else:
                        print(" 1과 2 중 하나를 입력하세요.\n")
                except ValueError:
                    print(" 숫자로 입력해주세요.\n")

            # 주문 리스트에 추가
            self.order_menu.append(f"{temp} {selected_menu}")
            self.order_price.append(selected_price)

            print(f" {temp} {selected_menu} : {selected_price}원 주문이 추가되었습니다.")

        # 주문이 끝난 후 주문표 출력
        self.table()
        # 주문이 끝난 후 결제 요청
        self.pay()

    def table(self):
        """주문표 출력"""
        if not self.order_menu:  # 주문이 없을 경우
            print("\n 주문 내역이 없습니다.")
            return

        max_length = max(len(item) for item in self.order_menu) + 10
        total_price = sum(self.order_price)

        # 외곽 테두리 상단
        print('\n' + ' ' * 2 + '@' + '@' * (max_length + 2) + '@')

        # 내부 여백 (위쪽)
        for _ in range(2):
            print(' ' * 2 + '@' + ' ' * (max_length + 2) + '@')

        # 주문 상품 출력
        for i in range(len(self.order_menu)):
            print(f"@ {self.order_menu[i]:<{max_length-8}} : {self.order_price[i]:>5}원 @")

        # 합계 금액 출력
        print(f"@ {'합계 금액':<{max_length-8}} : {total_price:>5}원 @")

        # 내부 여백 (아래쪽)
        for _ in range(2):
            print(' ' * 2 + '@' + ' ' * (max_length + 2) + '@')

        # 외곽 테두리 하단
        print(' ' * 2 + '@' + '@' * (max_length + 2) + '@')

    def pay(self):
        """지불"""
        total_price = sum(self.order_price)
        print(f"\n 총 결제 금액은 {total_price}원 입니다.")

        while True:  # 올바른 결제 방법이 선택될 때까지 반복
            try:
                pay_method = input("\n 결제 방법을 선택하세요 (1: 현금, 2: 카드) : ").strip().lower()

                if pay_method in ['1', 'cash']:  # 현금 결제
                    print(" 직원을 호출하겠습니다.")
                    break
                elif pay_method in ['2', 'card']:  # 카드 결제
                    print(" IC칩 방향에 맞게 카드를 꽂아주세요.")
                    break
                else:
                    print(" 다시 결제를 시도해 주세요.")  # 잘못된 입력 처리
            except Exception as e:
                print(f"오류 발생: {e}. 다시 시도해 주세요.")


a = Kiosk()
a.menu_print()  # 메뉴 출력
a.menu_select()
a.pay()
a.table()
```

```python
class Square:
    def __init__(self):
        """사용자가 원하는 도형을 선택하면 자동으로 해당 넓이 계산 메서드를 실행"""
        while True:
            try:
                self.square = int(input(
                    '넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n'
                    ' 1.직사각형 2.평행사변형 3.사다리꼴 \n >>> '
                ))

                if self.square == 1:
                    print('직사각형 함수는 rect()입니다.')
                    print(self.rect())  # 직사각형 넓이
                    break

                elif self.square == 2:
                    print('평행사변형 함수는 par()입니다.')
                    print(self.par())  # 평행사변형 넓이
                    break

                elif self.square == 3:
                    print('사다리꼴 함수는 trape()입니다.')
                    print(self.trape())  # 사다리꼴 넓이
                    break

                else:
                    print(' 1, 2, 3 중에서 다시 입력해주세요.\n')

            except ValueError:
                print(" 숫자를 입력해주세요.\n")

    def rect(self):  # 직사각형의 넓이
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>> ').split(','))
        area = width * vertical  # 넓이 계산
        return f'직사각형의 넓이는 : {area}'

    def par(self):  # 평행사변형 넓이
        base, height = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>> ').split(','))
        area = base * height  # 평행사변형 넓이 계산
        return f'평행사변형의 넓이는 : {area}'

    def trape(self):  # 사다리꼴의 넓이를
        lower, upper, height = map(int, input('밑변, 윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>> ').split(','))
        area = ((lower + upper) * height) / 2  # 사다리꼴 넓이 계산
        return f'사다리꼴의 넓이는 : {area}'

# 실행
a = Square()
```
