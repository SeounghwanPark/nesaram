```python
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

class Kiosk:
    def __init__(self):
        self.menu = menu  # ë©”ë‰´ ë¦¬ìŠ¤íŠ¸
        self.price = price  # ê°€ê²© ë¦¬ìŠ¤íŠ¸
        self.price_sum = 0  # ì„ íƒí•œ ìŒë£Œì˜ ê°€ê²©(ì´ˆê¸°ê°’)
        self.order_menu = []  # ì£¼ë¬¸ ë¦¬ìŠ¤íŠ¸
        self.order_price = []  # ê°€ê²© ë¦¬ìŠ¤íŠ¸

    def menu_print(self):
        """ë©”ë‰´ ì¶œë ¥"""
        print("\n ë©”ë‰´íŒ")
        for i in range(len(self.menu)):
            print(f"{i+1}. {self.menu[i]} : {self.price[i]}ì›")

    def menu_select(self):
        """ìŒë£Œ ì„ íƒ ë° ì£¼ë¬¸ ì¶”ê°€"""
        while True:
            print("\n ìŒë£Œë¥¼ ì„ íƒí•˜ì„¸ìš” (0 ì…ë ¥ ì‹œ ì£¼ë¬¸ ì¢…ë£Œ)")
            try:
                n = int(input("ìŒë£Œì˜ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš” : "))
            except ValueError:
                print(" ìˆ«ìë¡œ ì…ë ¥í•´ì£¼ì„¸ìš”.\n")
                continue

            # ì£¼ë¬¸ ì¢…ë£Œ
            if n == 0:
                print("\nğŸ›’ ì£¼ë¬¸ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
                break

            # ìœ íš¨í•œ ë©”ë‰´ ì„ íƒ í™•ì¸
            if 1 <= n <= len(self.menu):
                selected_menu = self.menu[n-1]
                selected_price = self.price[n-1]
            else:
                print(" ì—†ëŠ” ë©”ë‰´ì…ë‹ˆë‹¤. ë‹¤ì‹œ ì£¼ë¬¸í•´ ì£¼ì„¸ìš”.")
                continue

            # ìŒë£Œ ì˜¨ë„ ì„ íƒ
            t = 0
            while t not in [1, 2]:
                try:
                    t = int(input("HOT ìŒë£ŒëŠ” 1ì„, ICE ìŒë£ŒëŠ” 2ë¥¼ ì…ë ¥í•˜ì„¸ìš” : "))
                    if t == 1:
                        temp = "HOT"
                    elif t == 2:
                        temp = "ICE"
                    else:
                        print(" 1ê³¼ 2 ì¤‘ í•˜ë‚˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”.\n")
                except ValueError:
                    print(" ìˆ«ìë¡œ ì…ë ¥í•´ì£¼ì„¸ìš”.\n")

            # ì£¼ë¬¸ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
            self.order_menu.append(f"{temp} {selected_menu}")
            self.order_price.append(selected_price)

            print(f" {temp} {selected_menu} : {selected_price}ì› ì£¼ë¬¸ì´ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.")

        # ì£¼ë¬¸ì´ ëë‚œ í›„ ì£¼ë¬¸í‘œ ì¶œë ¥
        self.table()

    def table(self):
        """ì£¼ë¬¸í‘œ ì¶œë ¥"""
        if not self.order_menu:  # ì£¼ë¬¸ì´ ì—†ì„ ê²½ìš°
            print("\n ì£¼ë¬¸ ë‚´ì—­ì´ ì—†ìŠµë‹ˆë‹¤.")
            return

        max_length = max(len(item) for item in self.order_menu) + 10
        total_price = sum(self.order_price)

        # ì™¸ê³½ í…Œë‘ë¦¬ ìƒë‹¨
        print('\nâŸ' + '-' * max_length + 'âŸ')

        # ë‚´ë¶€ ì—¬ë°± (ìœ„ìª½)
        for _ in range(2):
            print('|' + ' ' * max_length + '|')

        # ì£¼ë¬¸ ìƒí’ˆ ì¶œë ¥
        for i in range(len(self.order_menu)):
            print(f"| {self.order_menu[i]:<{max_length-8}} : {self.order_price[i]:>5}ì› |")

        # í•©ê³„ ê¸ˆì•¡ ì¶œë ¥
        print(f"| {'í•©ê³„ ê¸ˆì•¡':<{max_length-8}} : {total_price:>5}ì› |")

        # ë‚´ë¶€ ì—¬ë°± (ì•„ë˜ìª½)
        for _ in range(2):
            print('|' + ' ' * max_length + '|')

        # ì™¸ê³½ í…Œë‘ë¦¬ í•˜ë‹¨
        print('âŸ' + '-' * max_length + 'âŸ\n')

    def pay(self):
        """ì§€ë¶ˆ"""
        total_price = sum(self.order_price)
        print(f"\nì´ ê²°ì œ ê¸ˆì•¡ì€ {total_price}ì› ì…ë‹ˆë‹¤.")
        # ì—¬ê¸°ì— ê²°ì œ ë¡œì§ ì¶”ê°€
        print("ê²°ì œê°€ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤. â˜• ê°ì‚¬í•©ë‹ˆë‹¤!")


a = Kiosk()  # ê°ì²´ ìƒì„±
a.menu_print()  # ë©”ë‰´ ì¶œë ¥
a.menu_select()
a.pay()  # ì§€ë¶ˆ
a.table()  # ì£¼ë¬¸í‘œ ì¶œë ¥
```

```python
# ì‚¬ê°í˜• ë„“ì´ë¥¼ êµ¬í•˜ëŠ” í´ë˜ìŠ¤ ì™„ì„±!
class Square:
    def __init__(self):
        self.square = int(input('ë„“ì´ë¥¼ êµ¬í•˜ê³  ì‹¶ì€ ì‚¬ê°í˜•ì˜ ìˆ«ìë¥¼ ì¨ì£¼ì„¸ìš”.\n 1.ì§ì‚¬ê°í˜• 2.í‰í–‰ì‚¬ë³€í˜• 3.ì‚¬ë‹¤ë¦¬ê¼´ \n >>>'))

        if self.square == 1:
            print('ì§ì‚¬ê°í˜• í•¨ìˆ˜ëŠ” rect()ì…ë‹ˆë‹¤.')  #Square í´ë˜ìŠ¤ì— rect() ë©”ì„œë“œë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤.

        elif self.square == 2:
            print('í‰í–‰ì‚¬ë³€í˜• í•¨ìˆ˜ëŠ” par()ì…ë‹ˆë‹¤.')    #Square í´ë˜ìŠ¤ì— par() ë©”ì„œë“œë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤.

        elif self.square == 3:
            print('ì‚¬ë‹¤ë¦¬ê¼´ í•¨ìˆ˜ëŠ” trape()ì…ë‹ˆë‹¤.')      #Square í´ë˜ìŠ¤ì— trape() ë©”ì„œë“œë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤.

        else:
            print('1, 2, 3 ì¤‘ì—ì„œ ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”')

    def rect(self):     #ì§ì‚¬ê°í˜•ì˜ ë„“ì´ë¥¼ êµ¬í•˜ëŠ” ë©”ì„œë“œ
        width, vertical = map(int, input('ê°€ë¡œ, ì„¸ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš”. ì˜ˆì‹œ : ê°€ë¡œ,ì„¸ë¡œ\n >>>').split(',')) #ê°€ë¡œ,ì„¸ë¡œ ì…ë ¥
        area = width * vertical    #ë„“ì´ë¥¼ êµ¬í•˜ëŠ” ì‹
        result = 'ì§ì‚¬ê°í˜•ì˜ ë„“ì´ëŠ” : ' + str(area)
        return result
    def par(self):     #í‰í–‰ì‚¬ë³€í˜• ë„“ì´êµ¬í•˜ëŠ” ë©”ì„œë“œ
        lower, height = map(int, input('ë°‘ë³€, ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”. ì˜ˆì‹œ : ë°‘ë³€,ë†’ì´\n >>>').split(','))  # ë°‘ë³€ê³¼ ë†’ì´ë¥¼ ì…ë ¥ë°›ìŒ
        area = lower * height  # í‰í–‰ì‚¬ë³€í˜• ë„“ì´ ê³„ì‚°
        result = 'í‰í–‰ì‚¬ë³€í˜•ì˜ ë„“ì´ëŠ” : ' + str(area)  # ê²°ê³¼ ë¬¸ìì—´ ìƒì„±
        return result  # ê²°ê³¼ ë°˜í™˜

    def trape(self):  # ì‚¬ë‹¤ë¦¬ê¼´ì˜ ë„“ì´ë¥¼ êµ¬í•˜ëŠ” ë©”ì„œë“œ
        lower, upper, height = map(int, input('ë°‘ë³€, ìœ—ë³€, ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”. ì˜ˆì‹œ : ë°‘ë³€,ìœ—ë³€,ë†’ì´\n >>>').split(','))  # ìœ—ë³€, ì•„ë«ë³€, ë†’ì´ ì…ë ¥ë°›ìŒ
        area = ((lower + upper) * height) / 2  # ì‚¬ë‹¤ë¦¬ê¼´ ë„“ì´ ê³„ì‚°
        result = 'ì‚¬ë‹¤ë¦¬ê¼´ì˜ ë„“ì´ëŠ” : ' + str(area)  # ê²°ê³¼ ë¬¸ìì—´ ìƒì„±
        return result
a = Square()
a.par()
a.trape()
```
