```python
data = pd.read_csv("../data/trip.csv")
# 검토할 컬럼 
columns_check = ['fare_amount', 'tolls_amount', 'tip_amount', 'passenger_count', 'trip_distance']

# 결측치 개수 및 비율 
missing_info = data[columns_check].isnull().sum()
missing_ratio = (missing_info / len(data)) * 100

# 데이터프레임
missing_df = pd.DataFrame({'Missing Count': missing_info, 'Missing Ratio (%)': missing_ratio})

print(missing_df)
```

```python
data = pd.read_csv("../data/trip.csv")
# 결측치 처리
data['fare_amount'].fillna(data['fare_amount'].median(), inplace=True)  # 중앙값으로 채우기

# 이상치 제거
newdata = data[data['fare_amount'] >= 0]  # 음수 요금 제거
newdata = newdata[newdata['passenger_count'] > 0]  # 승객 수 0명 제거
newdata = newdata[newdata['trip_distance'] > 0]  # 이동 거리 0 제거

# 중복 데이터 제거
newdata.drop_duplicates(inplace=True)

# 클리닝된 데이터
cleaned_newdata = newdata
print(cleaned_newdata)
```

```python
# 날짜 데이터를 datetime 형식으로
ver2_newdata['tpep_pickup_datetime'] = pd.to_datetime(ver2_newdata['tpep_pickup_datetime'])
ver2_newdata['tpep_dropoff_datetime'] = pd.to_datetime(ver2_newdata['tpep_dropoff_datetime'])
```

```python
# 주행 시간 (분)
ver2_newdata['trip_duration'] = (ver2_newdata['tpep_dropoff_datetime'] - ver2_newdata['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
# 1. 주행 시간 vs 주행 거리
sns.scatterplot(x=ver2_newdata['trip_duration'], y=ver2_newdata['trip_distance'])
# 주행시간이 길수록 주행거리가 증가하지만 일정시간이 지나면 거리증가가 둔화되어 교통체증가능성이 있으며, 
#일부 긴시간동안 거리가 짧은 경우는 교통체증(신호대기)의 정차시간이 포함될 가능성이 있음
# 특정 이동 거리 대비 시간이 과도하게 긴 경우는 이상 데이터 가능성있음. 
```

```python
# 2. 주행 시간 vs 요금
sns.scatterplot(x=ver2_newdata['trip_duration'], y=ver2_newdata['fare_amount'])
# 요금 예측모델을 할때 주행 시간이 중요한 변수로 적용할 가능성이 높음
# 너무 높거나 낮은 요금은 고객과의 협상(급행료로 더블등) 이거나 이상 데이터 가능성 있음
```

```python
# 3. 주행 거리 vs 요금
sns.scatterplot(x=ver2_newdata['trip_distance'], y=ver2_newdata['fare_amount'])
# 전반적으로 주행거리가 길수록 요금도 증가한다.
# 거리에 비해 요금이 과도하게 높거나 낮은 경우는 긴급료지불 또는 같은방행의 경우나 이상데이터 가능성 있음.
# 거리에 따른 평균 요금을 파악해서 요금 예측 모델 구축에 활용할수 있을거 같다.
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(ver2_newdata['trip_duration'], ver2_newdata['trip_distance'], ver2_newdata['fare_amount'], c=ver2_newdata['fare_amount'], cmap='viridis', alpha=0.6)
# 축 레이블
ax.set_xlabel("Trip Duration (minutes)")
ax.set_ylabel("Trip Distance (miles)")
ax.set_zlabel("Fare Amount ($)")
ax.set_title("3D Scatter Plot: Trip Duration, Trip Distance, and Fare Amount")
plt.show()
```

```python
# 요금 제한 
filtered_data = ver2_newdata[ver2_newdata['fare_amount'] <= 100]
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(filtered_data['trip_distance'], filtered_data['trip_duration'], filtered_data['fare_amount'], 
                 c=filtered_data['fare_amount'], cmap='plasma', alpha=0.7)
# 축 레이블
ax.set_xlabel("Trip Duration (minutes)")
ax.set_ylabel("Trip Distance (miles)")
ax.set_zlabel("Fare Amount ($)")
ax.set_title("3D Scatter Plot: Trip Duration, Trip Distance, and Fare Amount")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(filtered_data['fare_amount'], filtered_data['trip_distance'], filtered_data['trip_duration'], 
                 c=filtered_data['fare_amount'], cmap='plasma', alpha=0.7)
# 컬러바 (색상이 진할수록 요금이 높음)
cbar = plt.colorbar(sc, ax=ax, shrink=0.6)
cbar.set_label("Fare Amount ($)")
# 축 레이블
ax.set_xlabel("Fare Amount ($)")
ax.set_ylabel("Trip Distance (miles)")
ax.set_zlabel("Trip Duration (minutes)")
ax.set_title("3D Scatter Plot: Fare Amount, Trip Distance, and Trip Duration (Fare ≤ 100)")
plt.show()
```

```python
# 3D 산점도 (X축: 요금, Y축: 주행 거리, Z축: 주행 시간) - 컬러바를 파랑(낮은 요금) ~ 빨강(높은 요금)으로 설정
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(filtered_data['fare_amount'], filtered_data['trip_distance'], filtered_data['trip_duration'], 
                 c=filtered_data['fare_amount'], cmap='coolwarm', alpha=0.7)  
# 컬러바 
cbar = plt.colorbar(sc, ax=ax, shrink=0.6)
cbar.set_label("Fare Amount ($)")
# 축 레이블 
ax.set_xlabel("Fare Amount ($)")
ax.set_ylabel("Trip Distance (miles)")
ax.set_zlabel("Trip Duration (minutes)")
ax.set_title("3D Scatter Plot: Fare Amount, Trip Distance, and Trip Duration (Fare ≤ 100)")
plt.show()
```

```python
# 요금, 주행거리, 주행시간의 상관관계 로 봤을때 
# 택시 호출시 예상 도착시간을 예측 할 수 있을거 같으며, 요금이 80~100 사이의 값은 이상데이터 인지 궁금?
fare_values = ver2_newdata['fare_amount'].values
# 히스토그램 (요금 빈도)
plt.figure(figsize=(10, 6))
sns.histplot(fare_values, bins=50, kde=True, color='blue')
plt.xlabel("Fare Amount ($)")
plt.ylabel("Frequency")
plt.title("Pareto Chart for Taxi Fare Amounts")
# 두 번째 Y축 (누적 비율)
ax2 = plt.twinx()
sorted_fares = np.sort(fare_values)[::-1]  # 내림차순 정렬
cumulative_percentage = np.cumsum(sorted_fares) / np.sum(sorted_fares) * 100
# 누적 비율 라인 그래프 추가
ax2.plot(np.arange(len(sorted_fares)), cumulative_percentage, color='red', marker='o', markersize=2, linestyle='dashed')
ax2.set_ylabel("Cumulative Percentage (%)")
plt.show()
```
